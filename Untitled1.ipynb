```python
import numpy as np
import matplotlib.pyplot as plt
from components import *
%matplotlib widget
from matplotlib.patches import Polygon
plt.rcParams['figure.figsize'] = (6,3)
from scipy.spatial import ConvexHull
```

```python
grating = Grating.grating_from_file('config_pgm.ini')
mirror = Plane_Mirror.mirror_from_file('config_pgm.ini')
```

```python
a = grating.compute_corners()
```

```python
x, y, z = a[::1,0], a[::1,1], a[::1,2]
```

```python
vertices = np.array([z,y])
```

```python
vertices.T[::2]
```

```python
fig, ax = plt.subplots()
ax.scatter(a[:,2], a[:,1])
hull = ConvexHull(vertices.T[::2])
ax.fill(vertices.T[::2][hull.vertices,0],vertices.T[::2][hull.vertices,1])
ax.scatter(mirror.corners[::2,2], mirror.corners[::2,1])
```

```python
pgm = PGM(mirror=mirror, grating=grating)
```

```python
pgm.mirror.theta = pgm.theta
```

```python
pgm.mirror.compute_corners()
```

```python
fig,ax = plt.subplots()
pgm.draw(ax)
ax.set_ylim(-200,100)
ax.set_xlabel('x(mm)')
ax.set_ylabel('y(mm)')
plt.tight_layout()
ax.annotate(fr"$\alpha$ ={pgm.grating.alpha:.2f}$^\circ$" ,(-450,70))
ax.annotate(fr"$\beta$ ={pgm.grating.beta:.2f}$^\circ$" ,(-450,50))
ax.annotate(fr"$\theta$ ={pgm.theta:.2f}$^\circ$" ,(-450,30))
```
